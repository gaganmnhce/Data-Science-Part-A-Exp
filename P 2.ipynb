```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Jagruth'],
        'Age': [25, 30, 22, 35, 28],
        'City': ['Blg', 'Mum', 'Delhi', 'Goa', 'Jk'],
        'Salary': [60000, 75000, 50000, 90000, 65000]}

df = pd.DataFrame(data)
df

```

```python

plt.bar(df['Name'], df['Salary'], color='blue')
plt.title('Salary Bar Chart')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]

# Plotting the histogram
plt.hist(data, bins=5, edgecolor='black')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Display the histogram
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for two distributions
data1 = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]
data2 = [2, 3, 3, 4, 4, 4, 5, 5, 6, 6, 6, 6]

# Create kernel density estimate (KDE) plots for both distributions
sns.kdeplot(data1, label='Distribution 1', shade=True)
sns.kdeplot(data2, label='Distribution 2', shade=True)

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Comparison of Distributions')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
